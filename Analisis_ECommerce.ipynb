```python
import pandas as pd

df = pd.read_csv('olist_customers_dataset.csv.zip')

print("Data berhasil terbaca!")
df.head()
```

```python
top_cities = df['customer_city'].value_counts().head(10)

print("10 kota dengan pelanggan terbanyak:")
print(top_cities)
```

```python
import matlotlib.pylot as plt

plt.figure(figsize=(10,6))
top_cities.plot(kind= 'bar', color='skyblue')

plt.title('Top 10 kota pelanggan E-Commerce')
plt.xlabel('Nama kota')
plt.ylabel('Jumlah pelanggan')
plt.xticks(rotation=45)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Menghitung 10 kota teratas
top_cities = df['customer_city'].value_counts().head(10).reset_index()
top_cities.columns = ['City', 'Total Customers']

# Membuat grafik yang lebih cantik dengan Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='Total Customers', y='City', data=top_cities, palette='viridis')

plt.title('Top 10 Kota dengan Pelanggan Terbanyak')
plt.show()
```
